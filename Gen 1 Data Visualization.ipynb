## Data Visualization and Plotting

### Needed Libraries 

```python
import pickle
from statistics import mean, stdev, harmonic_mean
import shutil
import matplotlib.pyplot as plt
from tqdm import tqdm
```

### Plotting Function 

```python
def plot_avg_std(data1, data2, plot_title, x_axis_vals, fpp_rate):
    plt.figure(figsize=(8,5))
    x = range(1, len(data1) + 1)  # X-axis values
    avg1 = [item[0] for item in data1]  # Average values from tuples
    std1 = [item[1] for item in data1]  # Standard deviation values from tuples
    
    avg2 = [item[0] for item in data2]  # Average values from tuples
    std2 = [item[1] for item in data2]  # Standard deviation values from tuples
    
    # Plotting
    plt.errorbar(x, avg1, yerr=std1, fmt='o', capsize=4, color='Orange')
    plt.errorbar(x, avg2, yerr=std2, fmt='o', capsize=4, color='Green')
    plt.xlabel(f'Total of Transactions per Test', size=15)
    plt.ylabel('Time Needed per\n Operation (MiliSeconds)', size=15)

    plt.title(f'{plot_title}', size=20)
    
    xticks_values = x_axis_vals
    plt.xticks(x, xticks_values)
    plt.legend(['Cuckoo Filter', 'Bloom Filter'], loc=2)
    fig_title = ''.join(plot_title.split())
    file_name = fig_title+'_'+fpp_rate
    plt.savefig(f'{file_name}.png')
    plt.show()  
```

```python
unit = 1000
```

```python
with open('segments/inserts_001percent.pickle', 'rb') as b:
    loaded_insert_results = pickle.load(b)
    
with open('segments/query_bef_001percent.pickle', 'rb') as b:
    loaded_query_bef_results = pickle.load(b)
    
with open('segments/deletion_001percent.pickle', 'rb') as b:
    loaded_deletion_results = pickle.load(b)
    
with open('segments/query_aft_001percent.pickle', 'rb') as b:
    loaded_query_aft_results = pickle.load(b)
```

```python
bloom_avg_std_ins = [(mean(item)*unit,stdev(item)*unit) for num, item in enumerate(loaded_insert_results) if num%2==0] 
bloom_avg_std_bef = [(mean(item)*unit,stdev(item)*unit) for num, item in enumerate(loaded_query_bef_results) if num%2==0] 
bloom_avg_std_del = [(mean(item)*unit,stdev(item)*unit) for num, item in enumerate(loaded_deletion_results) if num%2==0] 
bloom_avg_std_aft = [(mean(item)*unit,stdev(item)*unit) for num, item in enumerate(loaded_query_aft_results) if num%2==0] 

cuckoo_avg_std_ins = [(mean(item)*unit,stdev(item)*unit) for num, item in enumerate(loaded_insert_results) if not num%2==0] 
cuckoo_avg_std_bef = [(mean(item)*unit,stdev(item)*unit) for num, item in enumerate(loaded_query_bef_results) if not num%2==0] 
cuckoo_avg_std_del = [(mean(item)*unit,stdev(item)*unit) for num, item in enumerate(loaded_deletion_results) if not num%2==0] 
cuckoo_avg_std_aft = [(mean(item)*unit,stdev(item)*unit) for num, item in enumerate(loaded_query_aft_results) if not num%2==0] 
```

```python
def mean_st_cal(list_tup):
    avg = []
    std = []
    for tup in list_tup:
        avg.append(tup[0])
        std.append(tup[1])
    print(mean(avg), mean(std))
```

```python
insert_ranges       = [25000  ,  50000, 100000, 200000,  400000,  800000, 1600000]
insert_ranges_ticks = ['25k',    '50k', '100k', '200k',  '400k',  '800k', '1600k']
query_del_ranges    = ['15k',    '30k',  '45k',  '60k',   '75k',   '90k', '105k', '120k', '135k', '150k'] 
```

```python
plot_avg_std(cuckoo_avg_std_ins, bloom_avg_std_ins, 'Insertion Performance', insert_ranges_ticks, '001percent_plot')
plot_avg_std(cuckoo_avg_std_bef, bloom_avg_std_bef, 'Query Performance\nBefore Deletions', query_del_ranges, '001percent_plot')
plot_avg_std(cuckoo_avg_std_del, bloom_avg_std_del, 'Deletion Performance', query_del_ranges, '001percent_plot')
plot_avg_std(cuckoo_avg_std_aft, bloom_avg_std_aft, 'Query Performance\nAfter Deletions', query_del_ranges, '001percent_plot')
```
